```python
# Jour 2 pour apprendre le code 

## Mes revenus mensuels prévus les 12 prochains mois 
revenus = [900, 900, 900, 1100, 1400, 1400, 1400, 1400, 2250, 2250, 2250, 2250]

print(f"Mes revenus par mois :", revenus)
print(f"Nombre de mois dans ma liste :", len(revenus) )
print(f"Mon revenu du mois 6 :", revenus[5])
# attention, ca commence a 0 !
```

```python
total_12_mois = sum(revenus)

print(f"En 12 mois je vais gagner :{total_12_mois} €")
print(f"Soit une moyenne de {round(total_12_mois / 12)}€ par mois")
```

```python
print("Evolution mois par mois :")
print("-" * 30)

for mois in range(1, 13) :
    revenu = revenus[mois-1]
    print(f"Mois {mois:2d} > {revenu} €")
```

```python
for mois in range(1, 13) :
    print(f"Mois {mois:2d} > {revenus[mois-1]} €") 
    print(f"Mois {mois:3d} > {revenus[mois-1]} €")
    print("---")
```
